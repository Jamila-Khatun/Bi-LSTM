```python
# Install required libraries
!pip install pandas numpy matplotlib seaborn scikit-learn tensorflow openpyxl imbalanced-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import RandomOverSampler
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
from google.colab import files
import io

# Set random seed for reproducibility
np.random.seed(42)

# ==============================================
# DATA LOADING AND VALIDATION
# ==============================================
print("╔════════════════════════════╗")
print("║   UPLOADING DATASET FILE   ║")
print("╚════════════════════════════╝")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_excel(io.BytesIO(uploaded[file_name]), header=None)

# Set column names and clean data
df.columns = ['text', 'label']
df = df.dropna()
df['text'] = df['text'].astype(str)

# Auto-detect actual classes in data
actual_classes = df['label'].unique()
print("\nDetected classes in your dataset:", actual_classes)
print("Class distribution:\n", df['label'].value_counts())

# Check if we have enough data per class
min_samples_per_class = 20
class_counts = df['label'].value_counts()
if any(class_counts < min_samples_per_class):
    print("\nWarning: Some classes have very few samples:")
    print(class_counts[class_counts < min_samples_per_class])

# ==============================================
# TEXT PREPROCESSING
# ==============================================
def preprocess_bangla_text(text):
    text = text.lower()
    # Keep Bangla characters, numbers, and basic punctuation
    text = ''.join(char for char in text if char.isalnum() or char.isspace() or
                  char in ['.', ',', '!', '?', '।', '্', 'া', 'ি', 'ী', 'ু', 'ূ', 'ৃ', 'ে', 'ৈ', 'ো', 'ৌ'])
    return text.strip()

df['text'] = df['text'].apply(preprocess_bangla_text)

# ==============================================
# HANDLE IMBALANCED DATA
# ==============================================
# Oversample minority classes
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(df[['text']], df['label'])
df = pd.DataFrame({'text': X_resampled['text'], 'label': y_resampled})
print("\nClass distribution after oversampling:\n", df['label'].value_counts())

# ==============================================
# DATA SPLITTING
# ==============================================
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'],
    test_size=0.2,
    stratify=df['label'],
    random_state=42
)

# ==============================================
# TEXT VECTORIZATION
# ==============================================
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
vocab_size = len(tokenizer.word_index) + 1

max_len = 100  # Reduced from 150 to prevent overfitting
X_train_pad = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=max_len)
X_test_pad = pad_sequences(tokenizer.texts_to_sequences(X_test), maxlen=max_len)

# ==============================================
# LABEL ENCODING
# ==============================================
label_map = {label: idx for idx, label in enumerate(df['label'].unique())}
inv_label_map = {v: k for k, v in label_map.items()}
num_classes = len(label_map)

y_train_enc = y_train.map(label_map).values
y_test_enc = y_test.map(label_map).values
y_train_cat = to_categorical(y_train_enc)
y_test_cat = to_categorical(y_test_enc)

# Compute class weights for imbalanced data
class_weights = compute_class_weight('balanced', classes=np.unique(y_train_enc), y=y_train_enc)
class_weights = dict(enumerate(class_weights))

# ==============================================
# MODEL ARCHITECTURE
# ==============================================
def build_model(vocab_size, num_classes):
    model = Sequential([
        Embedding(vocab_size, 128, input_length=max_len),  # Reduced from 256
        Bidirectional(LSTM(64, return_sequences=True)),    # Reduced from 128
        Dropout(0.3),                                     # Reduced from 0.5
        Bidirectional(LSTM(32)),                           # Reduced from 64
        Dropout(0.3),
        Dense(32, activation='relu'),                     # Reduced from 64
        Dense(num_classes, activation='softmax')
    ])

    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    return model

model = build_model(vocab_size, num_classes)
model.summary()

# ==============================================
# MODEL TRAINING
# ==============================================
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    X_train_pad, y_train_cat,
    epochs=20,  # Increased with early stopping
    batch_size=64,
    validation_split=0.1,
    class_weight=class_weights,
    callbacks=[early_stopping],
    verbose=1
)

# ==============================================
# EVALUATION
# ==============================================
def print_metrics(y_true, y_pred, classes):
    print("\n╔════════════════════════════╗")
    print("║        EVALUATION        ║")
    print("╚════════════════════════════╝")

    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

    # Print metrics
    print(f"\nAccuracy:  {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1 Score:  {f1:.4f}")

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=classes, zero_division=0))

# Predictions
y_pred = np.argmax(model.predict(X_test_pad), axis=1)
print_metrics(y_test_enc, y_pred, list(label_map.keys()))

# ==============================================
# CONFUSION MATRIX
# ==============================================
def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        cbar=False,
        annot_kws={'size': 12, 'weight': 'bold'}
    )

    plt.title('CONFUSION MATRIX\n', fontsize=16, pad=20, weight='bold')
    plt.xlabel('\nPREDICTED LABELS', fontsize=14, weight='bold')
    plt.ylabel('TRUE LABELS', fontsize=14, weight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)

    # Add borders
    for _, spine in plt.gca().spines.items():
        spine.set_visible(True)
        spine.set_linewidth(2)

    plt.tight_layout()
    plt.show()

plot_confusion_matrix(y_test_enc, y_pred, list(label_map.keys()))

# ==============================================
# CROSS-VALIDATION
# ==============================================
print("\n╔════════════════════════════╗")
print("║  5-FOLD CROSS VALIDATION  ║")
print("╚════════════════════════════╝")

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []

X_full = pad_sequences(tokenizer.texts_to_sequences(df['text']), maxlen=max_len)
y_full = df['label'].map(label_map).values

for fold, (train_idx, val_idx) in enumerate(skf.split(X_full, y_full)):
    print(f"\nFold {fold+1}/5:")

    # Build fresh model
    model_cv = build_model(vocab_size, num_classes)

    # Train with early stopping
    model_cv.fit(
        X_full[train_idx],
        to_categorical(y_full[train_idx]),
        epochs=20,
        batch_size=64,
        validation_split=0.1,
        class_weight=class_weights,
        callbacks=[early_stopping],
        verbose=1
    )

    # Evaluate
    y_pred_cv = np.argmax(model_cv.predict(X_full[val_idx]), axis=1)

    # Store results
    fold_results = {
        'fold': fold+1,
        'accuracy': accuracy_score(y_full[val_idx], y_pred_cv),
        'precision': precision_score(y_full[val_idx], y_pred_cv, average='weighted', zero_division=0),
        'recall': recall_score(y_full[val_idx], y_pred_cv, average='weighted', zero_division=0),
        'f1': f1_score(y_full[val_idx], y_pred_cv, average='weighted', zero_division=0)
    }
    cv_results.append(fold_results)

    print(f"Fold {fold+1} Results:")
    print(f"Accuracy: {fold_results['accuracy']:.4f}")
    print(f"Precision: {fold_results['precision']:.4f}")
    print(f"Recall: {fold_results['recall']:.4f}")
    print(f"F1 Score: {fold_results['f1']:.4f}")

# Print CV summary
cv_df = pd.DataFrame(cv_results)
print("\n╔════════════════════════════╗")
print("║ CROSS-VALIDATION SUMMARY  ║")
print("╚════════════════════════════╝")
print(cv_df.describe().loc[['mean', 'std']])
```
